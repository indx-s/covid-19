以下のコードのファイル名は、適時変更して下さい。

```python
df_covid19 = pd.read_csv("nhk_news_covid19_prefectures_daily_data.csv")
df_temp = df_covid19.copy()

# 日付以外の列を全て削除して日付だけのデータフレームを作成
# 列0 : 日付
# 列1 : 都道府県コード
# 列2 : 都道府県名
# 列3 : 各地の感染者数_1日ごとの発表数
# 列4 : 各地の感染者数_累計
# 列5 : 各地の死者数_1日ごとの発表数
# 列6 : 各地の死者数_累計

df_covid19.drop(df_covid19.columns[[1, 2, 3, 4, 5, 6]],axis=1,inplace=True)
df_covid19.drop_duplicates(inplace=True)


# 47都道府県の1日ごとの感染者数をdf_covid19に1列ずつmerge()

i = 1

while i < 48:
    df_pref = df_temp[df_temp["都道府県コード"]==i]
    # 列名:各地の感染者数_1日ごとの発表数を各都道府県の名前にrename()
    pref_name = df_pref.at[df_pref.index[0], '都道府県名']
    df_pref = df_pref.rename(columns={'各地の感染者数_1日ごとの発表数' : pref_name})
    df_pref.drop(df_pref.columns[[1, 2, 4, 5, 6]],axis=1,inplace=True)
    df_covid19 = pd.merge(df_covid19, df_pref, on='日付')
    i += 1
```

```python
# ファイルに出力
df_covid19.to_csv('df_covid19.csv',index=False, encoding='shift-jis')
from google.colab import files
files.download('df_covid19.csv')
```
